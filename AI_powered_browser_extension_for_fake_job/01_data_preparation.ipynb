```python
import pandas as pd

df1 = pd.read_csv('/content/drive/MyDrive/HireShield/data/fake_job_postings.csv')
df2 = pd.read_csv('/content/drive/MyDrive/HireShield/data/Fake Postings.csv')

print(df1.shape)
print(df2.shape)
```

```python
print("df1 columns:")
print(df1.columns.tolist())

print("\ndf2 columns:")
print(df2.columns.tolist())
```

```python
common_cols = [
    'title',
    'description',
    'requirements',
    'company_profile',
    'location',
    'salary_range',
    'employment_type',
    'industry',
    'benefits',
    'fraudulent'
]
```

```python
df1_aligned = df1[common_cols].copy()
df2_aligned = df2[common_cols].copy()
```

```python
merged_df = pd.concat([df1_aligned, df2_aligned], ignore_index=True)
```

```python
print("Merged shape:", merged_df.shape)
print("\nLabel distribution:")
print(merged_df['fraudulent'].value_counts())
```

```python
# Cell 4: Explore the merged dataset

print("="*60)
print("DATA EXPLORATION")
print("="*60)

# Basic info
print(f"\n Dataset Overview:")
print(f"   Total records: {len(merged_df):,}")
print(f"   Total columns: {len(merged_df.columns)}")
print(f"   Memory usage: {merged_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Check for missing values
print(f"\n Missing Values:")
missing = merged_df.isnull().sum()
missing_percent = (missing / len(merged_df) * 100).round(2)

for col in merged_df.columns:
    if missing[col] > 0:
        print(f"   {col}: {missing[col]:,} ({missing_percent[col]}%)")
    else:
        print(f"   {col}: 0 (no missing values)")

# Check data types
print(f"\n Data Types:")
print(merged_df.dtypes)

# Sample fake job
print(f"\n SAMPLE FAKE JOB:")
fake_sample = merged_df[merged_df['fraudulent'] == 1].iloc[0]
print(f"   Title: {fake_sample['title']}")
print(f"   Description: {fake_sample['description'][:200]}...")
print(f"   Location: {fake_sample['location']}")

# Sample real job
print(f"\n SAMPLE REAL JOB:")
real_sample = merged_df[merged_df['fraudulent'] == 0].iloc[0]
print(f"   Title: {real_sample['title']}")
print(f"   Description: {real_sample['description'][:200]}...")
print(f"   Location: {real_sample['location']}")

print(f"\n Exploration complete!")
```

```python
text_cols = [
    'title',
    'description',
    'requirements',
    'company_profile',
    'location',
    'salary_range',
    'employment_type',
    'industry',
    'benefits'
]

merged_df[text_cols] = merged_df[text_cols].fillna("")
```

```python
merged_df.to_csv(
    "/content/drive/MyDrive/HireShield/data/master_dataset.csv",
    index=False
)

print("Saved master_dataset.csv")
```
